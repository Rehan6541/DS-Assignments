2.)	A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


# 1.Business Problem

### 1.1 Objective
To determine whether there is a significant difference in the average Turn Around Time (TAT) of reports among the four laboratories. This helps the hospital evaluate and improve operational efficiency and identify laboratories with faster or slower reporting times.

### 1.2 Constraints

Ensure data accuracy and completeness.

Handle missing or outlier values appropriately.

Ensure statistical tests meet the required assumptions (normality and homogeneity of variance).

Maintain privacy and confidentiality of patient and laboratory data.

```python
#import necessory libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Load dataset
labtat_df = pd.read_csv('lab_tat_updated.csv')
labtat_df
```

# 2.Data Pre-processing

```python
# Remove rows with missing values
labtat_df.dropna(inplace=True)
```

```python
# Verify if missing values are removed
print(labtat_df.isnull().sum())
```

```python
# Display cleaned dataset
labtat_df.head()
```

```python
# Descriptive Statistics
labtat_df.describe()
```

```python
# Visualize distribution
plt.figure(figsize=(12, 6))
sns.histplot(labtat_df, kde=True)
plt.title('Distribution of TAT for Laboratories')
plt.show()
```

```python
# Boxplot to check for outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=labtat_df)
plt.title('Boxplot of TAT for Laboratories')
plt.show()
```

```python
# Remove outliers using IQR method
Q1 = labtat_df.quantile(0.25)
Q3 = labtat_df.quantile(0.75)
IQR = Q3 - Q1
labtat_df = labtat_df[~((labtat_df < (Q1 - 1.5 * IQR)) | (labtat_df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Verify dataset after outlier removal
labtat_df.describe()
```

```python
# Recheck for outliers after removal
plt.figure(figsize=(8, 6))
sns.boxplot(data=labtat_df)
plt.title('Boxplot After Outlier Removal')
plt.show()
```

```python
# Normality Test (Shapiro-Wilk Test)
shapiro_results = {col: stats.shapiro(labtat_df[col]) for col in labtat_df.columns}
for lab, result in shapiro_results.items():
    print(f"Shapiro-Wilk Test for {lab}:", result)
```

```python
# Homogeneity of Variance Test (Levene's Test)
levene_test = stats.levene(*[labtat_df[col] for col in labtat_df.columns])
print("Levene's Test for Equal Variance:", levene_test)
```

# 3.Model Building

```python
# 3.1 Partition the dataset
labs = [labtat_df[col] for col in labtat_df.columns]
labs
```

```python
# 3.2 Perform ANOVA Test (to compare means across multiple groups)
# Hypothesis:
# H0: There is no significant difference in average TAT among the laboratories
# H1: There is a significant difference in average TAT among the laboratories
def perform_anova(labs):
    anova_test = stats.f_oneway(*labs)
    print("ANOVA Test result:", anova_test)
    if anova_test.pvalue < 0.05:
        print("Reject Null Hypothesis: Significant difference exists among the laboratories.")
    else:
        print("Fail to Reject Null Hypothesis: No significant difference among the laboratories.")
```

```python
# 3.3 Model Improvement
# Ensure assumptions of ANOVA are met:
# 1. Normality (checked with Shapiro-Wilk Test)
# 2. Homogeneity of Variance (checked with Levene's Test)
if all(result.pvalue > 0.05 for result in shapiro_results.values()) and levene_test.pvalue > 0.05:
    print("Assumptions met. Performing ANOVA...")
    perform_anova(labs)
else:
    print("Assumptions not met. Consider using Kruskal-Wallis Test.")
```
